```python
p = pd.read_sql("SELECT * FROM people;", con=engine)
p.head(3)
```

```python
y=p.merge(salaries, left_on='playerid', right_on='playerid', how='left', indicator=False)
y.head(3)
```

```python
z=y.merge(college_playing, left_on='playerid', right_on='playerid', how='left', indicator=False)
z.head(3)
```

```python
zg=z.groupby(['playerid']).sum()
zg=zg.reset_index()
zg.head(3)
```

```python
zg = zg[zg.salary != 0.0]
zg.head(3)
```

```python
n=zg.merge(college_playing, left_on='playerid', right_on='playerid', how='left', indicator=False)
n.head(3)
```

```python
nn = n[n["schoolid"] == 'vandy']
nnn=nn.drop(columns=['yearid'])
nnnn=nnn.drop_duplicates()
nnnnn=nnnn.sort_values(by=['salary'], ascending=False)
nnnnn
```
